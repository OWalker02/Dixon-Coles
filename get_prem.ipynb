```python
import requests
from pprint import pprint as pp
import json
```

```python
with open('api_key.txt', 'r') as f:
    api_key = f.read()
```

```python
year = 2024
url = f'{url_base}/fixtures?league=39&season={year}'
headers = {
    'x-apisports-key': api_key
}

response = requests.get(url, headers=headers)
```
