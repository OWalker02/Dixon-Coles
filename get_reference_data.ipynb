```python
import requests
import json
```

```python
with open('api_key.txt', 'r') as f:
    api_key = f.read()
```

```python
# Base data to get
# timezone
# countries
# leagues
# leagues / seasons
```

```python
url_base = 'https://v3.football.api-sports.io'
```

```python
urls = {
    'timezone': f'{url_base}/timezone',   # List of 425 timezone strings
    'countries': f'{url_base}/countries',   # Every country the API has data for: alpha code (FR), flag url, name (France)
    'leagues': f'{url_base}/leagues',   # Full catalog of all competitions, each has: numeric ID, name, country, logo url, seasons (contains: coverage, current bool, start, end)
    'seasons': f'{url_base}/leagues/seasons'   # List of all seasons in API
}

# Example url for getting all currently active English comps: 
# https://v3.football.api-sports.io/leagues?country=England¤t=true
```

```python
headers = {
    'x-apisports-key': api_key
}
for cat, url in urls.items():
    response = requests.get(url, headers=headers)
    print(cat, url, len(response['response']))
    with open(f'data/reference/{cat}.json', 'w') as f:
        json.dump(response.json(), f)
```
